```python
import pandas as pd

# Sample DataFrame
data = {'category_name': ['A', 'B', 'A', 'B', 'C'],
        'quantity': [10, 15, 20, 25, 30]}
df = pd.DataFrame(data)

# Define a custom transformation function
def transform_quantity(row):
    if row['category_name'] == 'A':
        return row['quantity'] * 2  # Double quantity for category A
    elif row['category_name'] == 'B':
        return row['quantity'] * 1.5  # Increase quantity by 50% for category B
    else:
        return row['quantity']  # No transformation for other categories

# Apply the transformation to the 'quantity' column
df['transformed_quantity'] = df.apply(transform_quantity, axis=1)

# Display the result
print(df)
```
